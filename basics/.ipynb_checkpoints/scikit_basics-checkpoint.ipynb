# Sci kit learn's Estimator Api

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 10*rng.rand(50)
y = 2*x-1+rng.randn(50)
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
model
```

```python
X = x[:, np.newaxis]
X.shape
```

```python
model.fit(X,y)
```
